### Exercício 1

Escreva um programa que receba um inteiro n e calcule a seguinte série.

$\sum_{i=0}^{n}\frac{1}{2^i}$


````python
%%writefile ex1_aula1.cpp

#include <iostream>
#include <cmath>
#include <iomanip>      // std::setprecision
using std::cin; // usando apenas a função cin
using std::cout;

double somatorio(){
  int n;
  cin >> n;
  double resultado = 0;
  for (int i = 0; i < n ; i++){
    resultado += 1.0/pow(2, i);
  }
  
  return resultado;
}

double somatorio_v2_shift(){
  int n;
  cin >> n;
  double resultado = 0;
  for (int i = 0; i < n ; i++){
    resultado += 1.0/(1<<i);
  }
  
  return resultado;
}


double somatorio_v3_shift(){
  int n;
  cin >> n;
  double resultado = 1;
  int p = 1; // esse p facilita, pois ao invés de shiftar i vezes, ele shifyta uma vez do resultadp anterior. E ele inicia com 1, pois o primeiro shift de 0
  // é inútil, vai dar 1. Então já começa com i e itera n-1 vezes! ou seja, o i começa com 1
  for (int i = 0; i < n ; i++){
    resultado += 1.0/(p=p<<1);
  }
  
  return resultado;
}

int main(){
    double resultado_ex1 = somatorio();
    cout <<"Resultado do ex1 : " << std::setprecision(15) << resultado_ex1 << "\n" ;

    double resultado_ex1_v2 = somatorio_v2_shift();
    cout <<"Resultado do ex1 : " << std::setprecision(15) << resultado_ex1_v2 << "\n" ;

    double resultado_ex1_v3 = somatorio_v3_shift();
    cout <<"Resultado do ex1 : " << std::setprecision(15) << resultado_ex1_v3 << "\n" ;
    return 0;
}
````

### Exercício 2

Crie um programa que lê um número inteiro n e depois lê n números fracionários $x_{i}$
. Faça os seguintes cálculos e motre-os no terminal com 10 casas decimais.

$\bar{x} = \frac{1}{n}\sum_{i=1}^{n} x_i$

$ \sigma^2 = \frac{1}{n}\sum_{i=1}^{n} (x_i - \bar{x})^2 $

````python
%%writefile ex2_aula1.cpp

#include <iostream>
#include <cmath>
#include <iomanip>      // std::setprecision
#include <vector>
using std::cin; // usando apenas a função cin
using std::cout;
using std::vector;

vector<double> media_variancia(){
  int n;
  cout << "O seu vetor vai ter quantos números: ";
  cin >> n;
  vector <double> vec(n, 0.0);
  double media = 0;
  double variancia = 0;

  cout << "Escreva o(s) " << n <<  " números do seu vetor: ";
  for (int i = 0; i < n ; i++){
    cin >> vec[i]; 
    media += vec[i];
  }


  for (int i = 0; i < n ; i++){
    variancia += pow((vec[i]-(media/n)), 2);
  }

  return {media/n, variancia/n} ;
}

int main(){
    vector<double> resultado = media_variancia();
    cout <<"A média é : " << std::setprecision(10) << resultado[0] << "\n" ;
    cout <<"A variância é : " << std::setprecision(10) << resultado[1] << "\n" ;

    return 0;
}
````

### Exercício 3

Dados N pontos com coordenadas $(x_{i},y_{i})_{i=0}^{N}$, computar a matriz de distâncias D tal que:

$D_{i, j}$= Distância entre $(x_{i},y_{i}) e (x_{j},y_{j})$


Implemente um programa que calcule a matriz D acima. Sua entrada deverá estar no formato dos arquivos t6-in-*.txt e sua saída no formato dos arquivos t6-out-*.txt. Mostre as distâncias com 2 casas decimais.

Dicas:

a maneira mais fácil (não necessariamente a melhor) de alocar uma matriz é usando um vetor em que cada elemento é outro vetor.
faça uma implementação o mais simples possível. Vamos melhorá-la nas próximas tarefas.

````python
%%writefile ex3_aula1.cpp

#include <iostream>
#include <cmath>
#include <iomanip>      // std::setprecision
#include <vector>
#include <chrono>
using std::cin; // usando apenas a função cin
using std::cout;
using std::vector;
using namespace std::chrono;

vector<vector<double>> calcula_distancias_ingenuo(){
  int n;
  cin >> n;

  // Vetor resposta
  vector<vector <double>> resposta;

  // Vetor do x e y
  vector<double> x, y;

  // Primeiro vamos preencher os vetores de x e y
  for (int i = 0; i < n; i++){
    double xa, ya;
    cin >> xa >> ya;
    x.push_back(xa);
    y.push_back(ya);

  }

  for (int i = 0; i < n ; i++){
    vector<double>linha;
    for (int j = 0 ; j < n ; j ++){
      double x1 = pow((x[i] - x[j]), 2);
      double y1 = pow((y[i] - y[j]), 2);
      double distancia = sqrt(x1 + y1);
      linha.push_back(distancia);
    }
    resposta.push_back(linha);
  }

  for (int i = 0; i < n; i++){
      for (int j = 0; j < n; j++){
          cout << resposta[i][j] << " ";
      }
      cout << "\n";
  }


  return resposta;

}

int main(){
    auto start = high_resolution_clock::now(); // marca o inicio do tempo
    calcula_distancias_ingenuo();

    auto end = std::chrono::high_resolution_clock::now();
    std::chrono::duration<double, std::milli> elapsed = end - start;
    std::cout << "Tempo de execução: " << std::fixed << std::setprecision(2) << elapsed.count() << " ms\n";
    return 0;
}
````

Quando você dobra o tamanho da entrada (ou seja, de n para 2n), o tempo de execução do algoritmo pode aumentar em até um fator de 4 (ou seja, (2n)^2/n^2 = 4). No entanto, essa é uma generalização simplificada e nem sempre se aplica a todos os casos. 


**b)**Qual é a complexidade computacional de sua implementação?

A complexidade dessa função é O(n^2), pois ela tem dois loops aninhados que percorrem todo o vetor duas vezes, resultando em uma complexidade de O(n*n), ou seja, O(n^2). Além disso, dentro do loop interno, há algumas operações simples de soma, subtração e exponenciação, que têm complexidade constante e não afetam a complexidade geral da função.

### Por que o desempenho tão diferente?

As others have said, the issue is the store to the memory location in the array: x[i][j]. Here's a bit of insight why:

You have a 2-dimensional array, but memory in the computer is inherently 1-dimensional. So while you imagine your array like this:

0,0 | 0,1 | 0,2 | 0,3

----+-----+-----+----

1,0 | 1,1 | 1,2 | 1,3

----+-----+-----+----

2,0 | 2,1 | 2,2 | 2,3

Your computer stores it in memory as a single line:

0,0 | 0,1 | 0,2 | 0,3 | 1,0 | 1,1 | 1,2 | 1,3 | 2,0 | 2,1 | 2,2 | 2,3

In the 2nd example, you access the array by looping over the 2nd number first, i.e.:

```
x[0][0]
        x[0][1]
                x[0][2]
                        x[0][3]
                                x[1][0] etc...
```

Meaning that you're hitting them all in order. Now look at the 1st version. You're doing:

```
x[0][0]
                                x[1][0]
                                                                x[2][0]
        x[0][1]
                                        x[1][1] etc...
```
